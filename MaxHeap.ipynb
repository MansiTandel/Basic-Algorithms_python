```python
class Heap:
    
    def __init__(self):
        self._maxsize=10
        self._data=[-1] * self._maxsize
        #assign all elements the value -1.
        # total 10 elements
        self._csize=0
        # current size of heap
    def __len__(self):
        return len(self._data)
    
    def isempty(self):
        return len(self._data)==0
    
    def insert(self,ele):
        
        if self._csize==self._maxsize:
            print("No space available")
            return
        self._csize = self._csize + 1
        hi=self._csize
        
        while hi>1 and ele>self._data[hi//2]:
            # SWAP
            self._data[hi]=self._data[hi//2]
            # assigning parent to child
            hi=hi // 2
            # move to parent
        self._data[hi] = ele
        
        def max(self):
            if self._csize==0:
                print("empty heap")
                return
            return self._data[1]
        
        
    def deleteMax(self):
        if self._csize==0:
            print("empty heap")
            return
        # Replaceing the root not with last node
        ele=self._data[1]
        self._data[1]=self._data[self._csize]
        self._data[self._csize]=-1
        self._csize=self._csize-1
        
        i=1
        j=i*2
        
        while j<=self._csize:
            
            # check siblings and got the the bigger sibling by j+1
            if self._data[j]<self._data[j+1]:
                j=j+1
            
            # check parent with bigger sibling j
            # Swap if parent is smaller than its child j
            
            if self._data[i] < self._data[j]:
                temp=self._data[i]
                self._data[i]=self._data[j]
                self._data[j]=temp
                
                i=j
                j=i*2
                #now check its subtree down the heap and loop until done
            else:
                break
        return ele
        
S1 = Heap()
S1.insert(25)
S1.insert(14)
S1.insert(2)
S1.insert(20)
S1.insert(10)

print(S1._data)

S1.insert(40)
print(S1._data)

x=S1.deleteMax()
print(x)
print(S1._data)
            
y=S1.deleteMax()
print(y)
print(S1._data)
            
        
        
        
        
```
