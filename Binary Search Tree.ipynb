```python
class _Node:
    __slots__="_element","_left","_right"
    
    def __init__(self,element,left=None,right=None):
        self._element=element
        self._left=left
        self._right=right
        
class BinarySearchTree:
    
    def __init__(self):
        self._root=None
        
    def insert(self,troot,e):
        temp=None
        while troot:
            
            temp=troot
            
            if e==troot._element:
                return
            elif e<troot._element:
                troot=troot._left
            elif e>troot._element:
                troot=troot._right
           
        n = _Node(e)
        if self._root:
            if e < temp._element:
                temp._left = n
            else:
                temp._right = n
        else:
            self._root = n
            
    def rinsert(self,troot,e):
        if troot:
            if e < troot._element:
                troot._left=self.rinsert(troot._left,e)
            elif e > troot._element:
                troot._right=self.rinsert(troot._right,e)
        else:
            n = _Node(e)
            troot = n
        return troot
    
    def search(self,key):
        troot=self._root
        while troot:
            if key==troot._element:
                return True
            elif key < troot._element:
                troot=troot._left
            elif key > troot._element:
                troot=troot._right
        return False
                
    def rsearch(self,troot,key):  
        if troot:
            if key==troot._element:
                return True
            elif key < troot._element:
                return self.research(troot._left,key)
            elif key > troot._element:
                return self.research(troot._right,key)
        else:
            return False
        
    def delete(self,ele):
        p=self._root
        pp=None # parent of parent node
        
        # loop to find the element we are trying to delete
        # p ref the node we want to delete
        # pp references p's parent
        while p and p._element != ele:
            pp=p
            if ele < p._element:
                p=p._left
            else:
                p=p._right
        if not p:              #if the element is not found
            return False
        
        if p._left and p._right:
            s = p._left  # s is left subtree ref of p.
            ps = p
            while s._right:  # until we get the 
                             # largest element from left subtree
                ps=s
                s=s._right
            p._element=s._element
            p=s
            pp=ps
            
        c= None
        if p._left: # if just left subtree
            c = p._left
        else:      # if just right subtree
            c=p._right
        if p == self._root:  # check if only node
            self._root=None
        else:
            if p == pp._left: # check if its the left child of pp
                pp._left=c
            else:   # check if right child of pp
                pp._right=c
            
            
                
            
        
    
    def inorder(self,troot):
        if troot:
            self.inorder(troot._left)
            print(troot._element,end=" ")
            self.inorder(troot._right)
            
B1 = BinarySearchTree()
B1._root=B1.rinsert(B1._root,50)
B1.rinsert(B1._root,30)
B1.rinsert(B1._root,80)
B1.rinsert(B1._root,10)
B1.rinsert(B1._root,40)
B1.rinsert(B1._root,60)

B1.inorder(B1._root)
print()
print(B1.search(60))

print(B1.search(70))

print()
print(B1.rsearch(B1._root,50))

B1.delete(60)
B1.inorder(B1._root)



B1.delete(50)
B1.inorder(B1._root)

```
